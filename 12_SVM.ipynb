```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('breast_cancer.csv')
target = df['diagnosis']
s = set()
for val in target:
    s.add(val)
s = list(s)
x = df['radius_mean']
y = df['concavity_mean']
open_x = x[:50]
open_y = y[:50]
high_x = x[50:100]
high_y = y[50:100]
plt.figure(figsize=(10,8))
plt.scatter(open_x,open_y,marker='+',color='green')
plt.scatter(high_x,high_y,marker='_',color='red')
plt.show()
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
df = df.drop(['radius_mean','concavity_mean'],axis = 1)
Y = []
target = df['diagnosis']
for val in target:
    if val==1:
        Y.append(-1)
    else:
        Y.append(1)
df = df.drop(['diagnosis'],axis=1)
X = df.values.tolist()
X,Y = shuffle(X,Y)
xtr,xts,ytr,yts = [],[],[],[]
xtr,xts,ytr,yts = train_test_split(X,Y,train_size=0.9)
xtr,xts,ytr,yts = np.array(xtr),np.array(xts),np.array(ytr),np.array(yts)
ytr = ytr.reshape(len(ytr),1)
yts = yts.reshape(len(yts),1)
tr_f1 = xtr[:,0]
tr_f2 = xtr[:,1]
tr_f1 = tr_f1.reshape(512,1)
tr_f2 = tr_f2.reshape(512,1)
w1 = np.zeros((512,1))
w2 = np.zeros((512,1))
epochs = 1
alpha = 0.0001
while(epochs<10000):
    y = w1*tr_f1+w2*tr_f2
    pred = y*ytr
    print(epochs)
    count = 0
    for val in pred:
        if val>=1:
            cost=0
            w1 = w1 - alpha*(2*1/epochs*w1)
            w2 = w2 - alpha*(2*1/epochs*w2)
        else:
            cost = 1 - val
            w1 = w1 + alpha*(tr_f1[count]*ytr[count]-2*1/epochs*w1)
            w2 = w2 + alpha*(tr_f2[count]*ytr[count]-2*1/epochs*w2)
        count+=1
    epochs+=1

index = list(range(10,w1.shape[0]))
w1 = np.delete(w1,index)
w2 = np.delete(w2,index)
test_f1 = xts[:,0]
test_f2 = xts[:,1]
test_f1 = test_f1.reshape(len(test_f1),1)
test_f2 = test_f2.reshape(len(test_f2),1)
y_prd = w1[0]* test_f1+w2[0]+test_f2
w2
```
